##                                                                                      THE HINDU

The task assigned to you is to scrape archives from The Hindu website for years 2015 to 2021. You will be allotted tasks to scrape data based on different timeframes. The details to be scraped are given below:
1. Date: This column will contain the date on which news was reported.
2. Author: This column will contain name of the author of news report.
3. Vertical: This column will contain the news verticals (e.g. Entertainment, Sports, Lifestyle, Economy, etc.).
4. Headline: This column will contain the news headline.
5. Description: This column will contain the main content of the news report.

```python
#importing the necessary libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
```

```python
#connect to the driver
driver = webdriver.Chrome('chromedriver.exe')

#set delay
time.sleep(2)

#create website destimation
url = 'https://www.thehindu.com/'
driver.get(url)

#set delay
time.sleep(2)

#search_bar
search = driver.find_element_by_xpath('//input[@type="text"]').send_keys('Archives')

#set delay
time.sleep(1)

#click_button
search_button = driver.find_element_by_xpath('//button[@class="fa fa-search search-selection-icon"]').click()
```

```python
#click on archive
click_archive = driver.find_element_by_xpath('/html/body/div[1]/footer[2]/div[5]/div[1]/nav[2]/div[1]/ul/li[5]/a').click()

#set delay
time.sleep(1)

#click on the required month
click_month = driver.find_element_by_xpath('//*[@id="archiveWebContainer"]/div[2]/ul[11]/li[12]/a').click()
```

```python
#collect the Urls from Dates
Dates_Url = []

for d in driver.find_elements_by_xpath('//table[@class="archiveTable"]/tbody//tr/td/a'):
    Dates_Url.append(d.get_attribute('href'))

Dates_Url
```

```python
#click on the date to collect Articles Url
Date_Click = driver.find_element_by_xpath('//*[@id="archiveDayDatePicker"]/table/tbody/tr[1]/td[1]/a').click()
```

```python
#collect Articles Url

Articles_Url = []

for url in Dates_Url:
    driver.get(url)
    
    for a in driver.find_elements_by_xpath('//div[@class="col-lg-12 col-md-12 col-sm-12 col-xs-12"]/ul/li/a'):
        Articles_Url.append(a.get_attribute('href'))
    
Articles_Url
```

```python
#click on archive again
click_archive = driver.find_element_by_xpath('/html/body/div[2]/footer[2]/div[5]/div[1]/nav[2]/div[1]/ul/li[5]/a').click()

#set delay
time.sleep(2)

#click on the required month once again
click_month = driver.find_element_by_xpath('//*[@id="archiveWebContainer"]/div[2]/ul[11]/li[12]/a').click()

#set delay
time.sleep(1)

#click on the date to collect Articles Url once again
Date_Click = driver.find_element_by_xpath('//*[@id="archiveDayDatePicker"]/table/tbody/tr[1]/td[1]/a').click()

#set delay
time.sleep(1)

#click on the first Article
Click_Article = driver.find_element_by_xpath('//*[@id="section_1"]/div[2]/div/div/div/ul/li[1]/a').click()

#set delay
time.sleep(1)
```

```python
#create the empty databases
Date = []
Author = []
Vertical = []
Headline = []
Description = []

for url1 in Articles_Url:
    driver.get(url1)
    
    for e in driver.find_elements_by_xpath('/html/body/div[2]/div[2]/div/div/div/section/div/div/div/div[1]/div[2]/div/span[2]/none'):
        Date.append(e.text)
        
    for a in driver.find_elements_by_xpath('/html/body/div[2]/div[2]/div/div/div/section/div/div/div/div[1]/div[2]/span/a[2]'):
        Author.append(a.text)
    
    for v in driver.find_elements_by_xpath('/html/body/div[2]/div[1]/section/div/ul'):
        Vertical.append(v.text)

    for h in driver.find_elements_by_xpath('/html/body/div[2]/div[2]/div/div/div/section/div/div/div/div[1]/h1'):
        Headline.append(h.text)
        
    for i in driver.find_elements_by_xpath('//*[contains(@id,"content-body-")]'):
        Description.append(i.text)
```

```python
#create a dataframe
News = pd.DataFrame({})
News['Date'] = Date[0:10260]
News['Author'] = Author[0:10260]
News['Vertical'] = Vertical[0:10260]
News['Headline'] = Headline[0:10260]
News['Description'] = Description[0:10260]

News
```

```python
#saving the dataset in CSV format
News.to_csv('News_Hindu_Ver2')
```

```python
#saving the dataset in json format
News.to_json('Hindu_News_Ver2')
```
